```python
import json
import networkx as nx
from networkx.readwrite import json_graph
def read_json_file(filename):
    with open(filename) as f:
        js_graph = json.load(f)
    return json_graph.node_link_graph(js_graph)
```

```python
s1 = read_json_file("Прага.json")
```

```python
import pylab as plt
plt.figure(figsize=(30,10))
nx.draw(s1,with_labels=True,font_size=15,node_size=1000,edge_cmap=True,node_color='red')
plt.show()
```

Графы схем метро Тэгу2(урезанная версия схемы Тэгу) и метро Праги не изоморфны

```python
def necessary_isomorphic_condition(G, H):
    G_size = len(G.nodes())
    H_size = len(H.nodes())
    if(G_size != H_size):
        return False
    if sorted(list(dict(G.degree()).values())) != sorted(list(dict(H.degree()).values())):
        return False
    return True
```

```python
necessary_isomorphic_condition(s1,s4)
```
